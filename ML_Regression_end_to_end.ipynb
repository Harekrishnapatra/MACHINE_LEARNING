$Linear$ $Regression$

```python
#import Package
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
#Read the data
wine_df = pd.read_csv("winequality_red.csv")
wine_df
# Find the null value
wine_df.isnull().sum()
# split train test 
X = wine_df.drop("quality",axis=1)
y = wine_df["quality"]
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
# Devlop Model
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model
linear_model.fit(X_train,y_train)
y_pred = linear_model.predict(X_test)
y_pred
y_test

# Save the Model
# Method-1
import pickle
pickle.dump(linear_model,open("wine_df.pkl","wb"))
# Read the model 
pickle.load(open("wine_df.pkl","rb"))

# Method-2
import joblib
joblib.dump(linear_model,open("wine_df.joblib","wb"))
# Read the model
model_read = joblib.load(open("wine_df.pickle","rb"))
model_read
```

$Lasso$ $Regression$

```python
#import Package
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
#Read the data
wine_df = pd.read_csv("wine_df.csv")
wine_df
# Find the null value
wine_df.isnull().sum()
# split train test 
X = wine_df.drop("quality",axis=1)
y = wine_df["quality"]
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
# Devlop Model
from sklearn.linear_model import Lasso
lasso_model = Lasso()
lasso_model
lasso_model.fit(X_train,y_train)
y_pred = lasso_model.predict(X_test)
y_pred
y_test

# MSE

from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_pred,y_test)
print("Mean Square Error",mse)

# Save the Model
# Method-1
import pickle
pickle.dump(lasso_model,open("wine_df.pkl","wb"))
# Read the model 
pickle.load(open("wine_df.pkl","rb"))

# Method-2
import joblib
joblib.dump(lasso_model,open("wine_df.joblib","wb"))
# Read the model
model_read = joblib.load(open("wine_df.pickle","rb"))
model_read
```

$Elastic$ $Net$

```python
#import Package
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import train_test_split
#Read the data
wine_df = pd.read_csv("wine_df.csv")
wine_df
# Find the null value
wine_df.isnull().sum()
# split train test 
X = wine_df.drop("quality",axis=1)
y = wine_df["quality"]
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
# Devlop Model
from sklearn.linear_model import ElasticNet
elastic_net = ElasticNet(alpha=0.1,l1_ratio=0.5)
elastic_net
elastic_net.fit(X_train,y_train)
y_pred = elastic_net.predict(X_test)
y_pred
y_test

# MSE

from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_pred,y_test)
print("Mean Square Error",mse)

# Save the Model
# Method-1
import pickle
pickle.dump(elastic_net,open("wine_df.pkl","wb"))
# Read the model 
pickle.load(open("wine_df.pkl","rb"))

# Method-2
import joblib
joblib.dump(elastic_net,open("wine_df.joblib","wb"))
# Read the model
model_read = joblib.load(open("wine_df.pickle","rb"))
model_read
```
